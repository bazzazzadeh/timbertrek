```python
from json import load
import pandas as pd
import timbertrek
```

```python
# Load the trie and dataset files
trie = load(open("./car_0.15_0.015.json", "r"))
data_df = pd.read_csv("./car_data.csv")
feature_names = data_df.columns[:-1].tolist()

feature_description = {
    'buying': {
        'info': 'Buying price',
        'type': 'is',
        'short': 'Buying price'
    },
    'maint': {
        'info': 'Maint price',
        'type': 'is',
        'short': 'Maintenance price'
    },
    'doors': {
        'info': 'Number of doors',
        'type': 'is',
        'short': 'Doors'
    },
    'persons': {
        'info': 'Persons capacity',
        'type': 'is',
        'short': 'Persons'
    },
    'lug_boot': {
        'info': 'Luggage boot size',
        'type': 'is',
        'short': 'Luggage size'
    },
    'safety': {
        'info': 'Estimated safety',
        'type': 'is',
        'short': 'Safety'
    }
}

decision_paths = timbertrek.transform_trie_to_rules(
    trie,
    data_df,
    feature_names=feature_names,
    feature_description=feature_description,
)
```

```python
timbertrek.visualize(decision_paths, width=430, height=550)
```
